### Which Department has the highest or lowest salary / Which department pays high

###### 1. Using the countplot method of seaborn

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Department',hue='salary',data=data)
plt.xticks(rotation=90)
plt.show()
```

###### 2. Using the Groupby Method

```python
salary_data=data.groupby(['Department','salary'])['salary'].count()
```

```python
salary_data
```

### How many employees left from the perticular department and from which department highest employees has been left 

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Department',hue='left',data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
left_data=data.groupby(['Department','left'])['left'].count()
```

```python
data['Department'].value_counts()
```

```python
left_data.unstack()
```
